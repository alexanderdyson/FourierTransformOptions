# Transform Methods and Option Pricing

Transform methods provide a computationally efficient alternative for pricing options when the underlying asset follows a stochastic process with a known characteristic function. If a random variable has a probability density function (PDF), then its characteristic function is the Fourier transform of that PDF.

The Inversion Theorem is the fundamental result linking these two representations: given the characteristic function, we can recover the probability distribution via an inverse Fourier transform.



Definition:

The characteristic function $ϕ(ν)$ of a distribution, $X$, is the Fourier transform of the probability density function $f(x)$,

$$\phi (\nu) = \int_{-\infty}^{\infty} {e^{iν x}}f(x) \, dx \\[1em]$$



$$= \mathbb{E} \left({e^{iν x}} \right)$$

* With  $\phi (ν) $, the function $f(x)$ can be recovered via the inverse Fourier transform,

$$f(x) =\frac{1}{2\pi} \int_{-\infty}^{\infty} {e^{-iν x}} \phi(ν) dν \\[1em]$$

## Carr and Madan Approach

Carr and Madan (1999) developed an FFT-based method for pricing options across multiple strikes simultaneously. The approach begins by constructing the Fourier transform of a modified call option price with respect to the logarithmic strike $k$, then recovers the PDF from the characteristic function via the inverse Fourier transform. The option price is then obtained by integrating the product of the payoff with the PDF.

Set-up:

* $S_{T}$: $\,$ $T$-time price

* $f(S_{T}|S_{0})$: $\,$ density of $S_{T}$

* $q(s_{t}|s_{0})$: $\,$ density of $s_{T} = {\ln}(S_{T})$

* $k = {\ln}(K)$: $\,$ log of the strike price

* $\phi (ν)$: $\,$ CF of log stock price process

* $C_{T}(k)$: $\,$ price of a $T$-maturity call with $K={e^{k}}$

Formulation:
* The European call option price $C_{T}(k)$ can be expressed as,



$${e^{-rT}}\mathbb{E_{0}} [(S_{T}-K)^+] =  {e^{-rT}}  \int_{K}^{\infty} (S_{T}-K) f \,(S_{T}) \, dS_{T} $$

* Change of variables into the log space,

$$ = {e^{-rT}}  \int_{K}^{\infty}  \,(e^{s_{T}}-e^{k}) \, q \, (s_{T})\, ds_{T} $$

$$ = C_{T}(k) $$

From the martingale property, $\space \mathbb{E^{ℚ}}[S_{T}]=S_{0}e^{rT}$, it is known that $\space \lim_{k \to -\infty}C(k) = S_{0}e^{rT}$ does not converge to zero. This integrability condition means the Fourier transform does not exist. To fix this, an exponential damping factor $\space e^{αk} \space$ with $\space α>0$ makes it possible to integrate $C(k)$.

Modified Call and its Fourier Transform:

* Define, $ c_{T}(k) = e^{{αk}}C_{T}(k) $. Let $ \Psi_{T}(ν) $ be its Fourier transform,

$$ \Psi_{T}(\nu) = \int_{-\infty}^{\infty} e^{ivk}c_{T}  (k) \space dk $$

$$ \Psi_{T}(\nu) = \int_{-\infty}^{\infty} e^{ivk}e^{αk}C_{T} (k) \space dk $$

* Now use the inverse Fourier transform to get,

$$ C_{T}(k) = \frac{e^{-αk}}{2\pi} \int_{-\infty}^{\infty} \, e^{-ivk} \, {\Psi_{T}}(ν) \, dν $$

The goal is to obtain $ {\Psi_{T}}(\nu) $ analytically. We evaluate by switching the order of integration,

$$ \Psi_{T}(ν) = \int_{-\infty}^{\infty} e^{ivk}c_{T} (k) \space dk $$




Substituting $ e^{α k} $ and $ \Bigl ( e^{-r T} \int_{k}^{\infty} (e^{s}-e^{k}) q  (s)  ds \Bigl )$ for $ c_{T} $ and $ (k)$ respectively,



$$ = \int_{-\infty}^{\infty} e^{ivk} e^{α k} \Bigl ( e^{-r T} \int_{k}^{\infty} (e^{s}-e^{k}) q  (s)  ds \Bigl ) dk $$

By Fubini's Theorem, if a function of two variables is absolutely integrable over a region, then its iterated integrals and its double integral over the region are all equal. Since $ q(s) $ does not depend on $ k $, we can perform the integration with respect to $ k $ first,



$$ = e^{-r T} \int_{-\infty}^{\infty} \int_{-\infty}^{s} e^{(α+iν)k} (e^{s}-e^{k}) \space q (s) \space dk ds $$

$$ = e^{-r T} \int_{-\infty}^{\infty} q(s) \Bigl (\int_{-\infty}^{s} e^{(α+iν)k} (e^{s}-e^{k}) dk \Bigl )  $$

The inner integral evaluates to:

$$ = e^{s} \frac{e^{(α+iν)k}}{(α+iν)} \Big|_{-\infty}^{s} - \frac{e^{(α+iν+1)k}}{(α+iν+1)} \Big|_{-\infty}^{s} $$

$$ = e^{s} \frac{e^{(α+iν)s}}{(α+iν)} - \frac{e^{(α+iν+1)s}}{(α+iν+1)} $$

$$ = \frac{e^{(α+iν+1)s}}{(α+iν)(α+iν+1)} $$

Substituting back into the outer integral and recognizing the characteristic function,


$$ \Psi_{T}(ν) = e^{-rT} \int_{-\infty}^{\infty} q(s) \frac{e^{(α+iν+1)s}}{(α+iν)(α+iν+1)} ds$$

$$= \frac{e^{-rT}}{(α+iν)(α+iν+1)} \int_{-\infty}^{\infty} e^{(\alpha+iν+1)s}q(s)ds $$

$$ = \frac{e^{-rT}}{(α+iν)(α+iν+1)} ϕ(ν-(α+1)i) $$

* We know the characteristic function of the log of the stock price, $ ϕ(ν) = \int {e^{iνs}}q(s)ds $. The expression above shows that this characteristic function is simply being evaluated at $(\nu-(α+1)i)$.

* The Fourier transform $ \Psi_{T}(ν)$ of the modified call can be obtained analytically as long as the characteristic function of the log stock price $ϕ(ν)$ is known.

Recall that the inverse Fourier transform, multiplied by the reciprocal of the exponential factor, yields the undamped call prices,



$$ C_{T}(k) = \frac{e^{-αk}}{2π} \int_{-\infty}^{\infty} e^{-ivk}\Psi_{T}(ν)dν$$

Where,

$$ \Psi_{T}(ν) = \frac{e^{-rT}}{(α+iν)(α+iν+1)} ϕ(ν-(α+1)i)$$

## Integral Evaluation

Given $ϕ(ν)$, the inverse Fourier transform multiplied by the reciprocal of the exponential factor gives us the undamped call prices. Next we need to evaluate a definite integral. Rather than integrating to infinity, we truncate the upper limit to $B$. Since the integrand has an even real part, we can reduce the domain from $ \frac{1}{2π} \int_{-\infty}^{+\infty} $ to $ \frac{1}{π} \int_{0}^{\infty} $.


$$ C_{T}(k) = \frac{e^{-αk}}{2π} \int_{-\infty}^{\infty} e^{-ivk}\Psi_{T}(ν)dν$$

$$ C_{T}(k) = \frac{e^{-αk}}{π} \int_{0}^{\infty} e^{-ivk}\Psi_{T}(ν)dν$$

Truncating at $B$,

$$ ≈ \frac{e^{-αk}}{π} \int_{0}^{B} e^{-ivk}\Psi_{T}(ν)dν $$

Set the number of summands $N = 2^n$ to utilize the computational efficiency of the FFT, and choose the grid spacing $η$ within the Fourier domain. We divide the interval $[0, B]$ into $N$ equal sub-intervals of length $η$ where,

$$ \eta = \frac{B}{N} $$




The grid points from $0$ to $B$ are,


$$ ν_{j} = (j-1)η $$ for $$ j = 1,...,N+1 $$

Decomposing into $N$ sub-integrals,

$$ C_{T}(k) \approx \frac{e^{-αk}}{π} \sum_{j=1}^{N} \int_{ν_{j}}^{ν_{j+1}} e^{-ivk}\Psi_{T}(ν)dν $$

Applying the trapezoidal rule to approximate each sub-integral,

$$ \int_{ν_{j}}^{ν_{j+1}} e^{-ivk}\Psi_{T}(ν)dν \space ≈ \space \frac{η}{2} \Bigl (e^{-iν_{j}k} Ψ_{T}(\nu_{j})+e^{-iν_{j+1}k} Ψ_{T}(ν_{j+1}) \Bigl) $$

Substituting,

$$ C_{T}(k) \approx \frac{e^{-αk}}{π} \sum_{j=1}^{N} \int_{ν_{j}}^{ν_{j+1}} e^{-ivk}\Psi_{T}(ν)dν   $$

$$ C_{T}(k) \approx \frac{e^{-αk}}{π} \sum_{j=1}^{N} \space \frac{η}{2} \Bigl (e^{-iν_{j}k} Ψ_{T}(\nu_{j})+e^{-iν_{j+1}k} Ψ_{T}(ν_{j+1}) \Bigl) $$

This gives us a call price for a specific strike,

$$ C_{T}(k) \approx \frac{e^{-αk}}{π} \sum_{j=1}^{N} e^{-iν_{j}k} Ψ_{T}\space(ν_{j})\space w_{j}$$

Where,

$$ w_{j} = \begin{Bmatrix} \frac{1}{2}η & j=1\\ η & \text{otherwise} \end{Bmatrix} $$

## Fast Fourier Transform

The Discrete Fourier Transform can be computed using the Fast Fourier Transform algorithm in $O(N \log_{2} N)$ time, yielding an entire range of option prices for multiple strikes in a single pass. Below we walk through the setup and parameters for the algorithm.

For, $$ m = 1,...,N $$


$$ C_{T}(k_{m}) \approx \frac{e^{-αk_{m}}}{π} \sum_{j=1}^{N} e^{-iν_{j}k_{m}} Ψ_{T}\space(ν_{j})\space w_{j}$$

$$ C_{T}(k_{m}) \approx \frac{e^{-αk_{m}}}{π} \sum_{j=1}^{N} e^{-i(j-1)η(m-1)Δk} e^{-iβv_{j}} Ψ _{T}(\nu_{j})w_{j} $$

We can write $ k_{m} = β+(m-1)Δk  $, where:
* $β$ is the starting point for the first log-strike, chosen by the user
* $Δk$ is the step size in the log-strike space
* For simplicity, we write $Δk$ as $λ$

$$ C_{T}(k_{m}) \approx \frac{e^{-αk_{m}}}{π} \sum_{j=1}^{N} e^{-i(j-1)η(m-1)Δk} e^{-iβv_{j}} Ψ _{T}(v_{j})w_{j} $$

For simplicity, define $ \space x(j) = e^{-iβv_{j}} Ψ _{T}(v_{j})w_{j} $

$$ C_{T}(k_{m}) \approx \frac{e^{-αk_{m}}}{π} \sum_{j=1}^{N} e^{-iλη (j-1)(m-1)} x(j)$$

The FFT provides an efficient algorithm for computing this sum. Setting $λη = \frac{2π}{N}$,

$$ ω(m) = \sum_{j=1}^{N} e^{-i \frac{2π}{N} (j-1)(m-1)} x(j) $$

For, $$ m = 1,...,N $$

## Implementation of FFT

Given $ ϕ(\nu) $, the characteristic function of the log stock price process,

Choose:
  * $η$ (Fourier grid spacing)
  * $N=2^n$ (number of grid points)
  * $β$ (starting log-strike)

Calculate:
  * $λ=\frac{2π}{Nη}$
  * $ν_{j}=(j-1)η$
  * Set $α>0$



Form vector $x$,

$$ x = \begin{pmatrix} x_{1} \\ x_{2} \\ \vdots \\ x_{N} \end{pmatrix} = \begin{pmatrix} \frac{η}{2}\frac{e^{-rT}}{(α+iν_{1})(α+iν_{1}+1)} e^{-iβv_{1}} \phi(ν_{1}-(α+1)i) \\ \eta \frac{e^{-rT}}{(α+iν_{2})(α+iν_{2}+1)} e^{-iβv_{2}} \phi(ν_{2}-(α+1)i) \\ \vdots \\ \eta \frac{e^{-rT}}{(α+iν_{N})(α+iν_{N}+1)} e^{-iβv_{N}} \phi(ν_{N}-(α+1)i) \end{pmatrix}  $$

Pass this vector through the FFT routine,

$$ y = \text{fft}(x) $$

The call prices at strike $k_{m}$ for $m=1,...,N$ are,

$$ \begin{pmatrix} C_{T}(k_{1}) \\ C_{T}(k_{2}) \\ \vdots \\ C_{T}(k_{N}) \end{pmatrix} = \begin{pmatrix} \frac{e^{-αk_{1}}}{π} \text{Re}(y_{1})\\ \frac{e^{-αk_{2}}}{π} \text{Re}(y_{2})\\ \vdots \\
\frac{e^{-αk_{N}}}{π} \text{Re}(y_{N})\\ \end{pmatrix}  $$

Where, $$k_{m} = β+(m-1)λ$$

This produces an output vector of $N$ call option prices for the corresponding strikes, with maturity fixed.


## Recap and Choice of Parameters

$$ k_{m} = β+(m-1)λ \space \space \text{for} \space m=1,...,N $$

$β$ determines the first log-strike. Two common choices:

* **Center the grid at-the-money:** set $β=\ln(S_{0})-\frac{N}{2}λ$

* **Start at a specific strike $K$:** set $β=\ln(K)$
